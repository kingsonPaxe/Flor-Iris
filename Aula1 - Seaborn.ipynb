# Visualização de dados

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
dt = pd.DataFrame(
    data.data, 
    columns=data.feature_names
    )
```

```python
dt.shape
```

```python
dt.head(5)
```

```python
dt.tail()
```

```python
x = dt["sepal width (cm)"]
y = dt["sepal length (cm)"]
sns.scatterplot(dt,x = x, y=y)
```

```python
sns.scatterplot(x = dt["petal width (cm)"], y = dt["petal length (cm)"])
```

```python
sns.scatterplot(x = dt["sepal width (cm)"], y = dt["petal length (cm)"])
```

```python
sns.histplot(dt["sepal width (cm)"])
```
